```python
#Bunch of methods since I end up using the same commands repeatedly.

def createdatabase(name):
    mycursor = db.cursor()
    mycursor.execute('CREATE DATABASE {}'.format(name))

def createtable(name):
    mycursor = db.cursor()
    mycursor.execute("CREATE TABLE {} (date DATE, num FLOAT)".format(name))
    db.commit()

def addrow(name, day, value):
    mycursor = db.cursor()
    mycursor.execute("INSERT INTO {} VALUES('{}',{})".format(name, day, value))
    db.commit()

def deletetable(name):
    mycursor = db.cursor()
    mycursor.execute('DROP TABLE {};'.format(name))
    db.commit()

def showtable(name):
    mycursor = db.cursor()
    mycursor.execute("SELECT * FROM {}".format(name))

    for x in mycursor:
        print(x)

def createschema(name):
    mycursor = db.cursor()
    mycursor=db.cursor('CREATE SCHEMA {}'.format(name))
    db.commit()

def showtables():
    mycursor = db.cursor()
    mycursor.execute('SHOW TABLES;')
    result = mycursor.fetchall()

    for x in result:
        print(x)

def readtable(name):
    mycursor = db.cursor()
    mycursor.execute("SELECT * FROM {}".format(name))

    for x in mycursor:
        print(x)
```
